```python
temp = df.groupby(['Weekday','Hour']).count()['Invoice_No'].reset_index().pivot( index='Weekday' , columns='Hour' , values ='Invoice_No'  )


plt.figure(figsize=(12,6))
sns.heatmap(temp , cmap = 'YlGn' )
plt.title("Frequency of Day Vs Hour of day")
plt.show()
```
