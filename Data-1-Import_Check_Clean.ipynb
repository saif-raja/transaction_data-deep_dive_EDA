# dealing with NULL values in Cust_ID

ignoring NULLs in Product Description as it is a non Critical column 

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df [ df['Cust_ID'].isna() ] #.to_clipboard()
```


can impute customer ID from other matching Invoice Numbers 

as there is a one way distinct mapping ( one to many ) for customers to invoices 

Ie one customer can have many invoices , but one invoice is associated with only a single customer

so can use Repeating Invoice Numbers to get Customer ID

need to create a Customer-Invoice Map , and then use that to impute teh missing NaNs

```python
df['Cust_ID'].isna().sum()
```

```python
invoice_customer_mapping = df[['Invoice_No','Cust_ID']] \
                            .dropna( how = 'any' ) \
                            .groupby( by = 'Invoice_No').agg(lambda x : x.value_counts().index[0] ) \
                            .to_dict( 'dict' )['Cust_ID']

df['Cust_ID'] = df['Cust_ID'].fillna( df['Invoice_No'].map(invoice_customer_mapping) )
```

```python
invoice_customer_mapping = df[['Invoice_No','Cust_ID']] \
                            .dropna( how = 'any' ) \
                            .groupby( by = 'Invoice_No').agg(pd.Series.mode) \
                            .to_dict( 'dict' )['Cust_ID']

df['Cust_ID'] = df['Cust_ID'].fillna( df['Invoice_No'].map(invoice_customer_mapping) )
```

# Final Level Fix ( Aggergation )

```python
df['Sale_Value'] = df['Product_Qty'] * df['Product_Price'] 
```

```python
df
```

```python
final_columns = [
'Invoice_No',
'Product_Code',
'Cust_ID',
'Invoice_Date',
'Invoice_Status',
'Product_Qty',
'Product_Price',
'Sale_Value',
'Misc_Product_Code_Flag',
'Cust_Country',
'Product_Desc',
]

# mode = lambda x : x.value_counts().index[0]
# mode = lambda x : x.mode(axis=0, numeric_only=False, dropna=False)

# mode did not work , because of pandas doesnt have a implementation , and the hacks neede dtoo much time to compute 
# used max instead , because anyways the columns have a single value , so mode and max would return the same values

final_df = df[final_columns].groupby(['Invoice_No','Product_Code']).agg(
    {
        'Cust_ID'                   : 'max' ,
        'Invoice_Date'              : 'min' ,
        'Invoice_Status'            : 'max' ,
        'Product_Qty'               : 'sum' ,
        'Product_Price'             : 'mean',
        'Sale_Value'                : 'sum' ,
        'Misc_Product_Code_Flag'    : 'max' ,
        'Cust_Country'              : 'max' ,
        'Product_Desc'              : 'max' ,
    }
)

final_df = final_df.reset_index()
final_df
```

Next we also need to create some Auxillary columns for doing Time Dimension analysis


```python
df['Invoice_Date']  = df['Invoice_Date'].astype('datetime64[ns]')
df["Year"]          = df['Invoice_Date'].dt.year
df["Quarter"]       = df['Invoice_Date'].dt.quarter
df["Month"]         = df['Invoice_Date'].dt.month
df["Week"]          = df['Invoice_Date'].dt.week
df["Weekday"]       = df['Invoice_Date'].dt.weekday
df["Day"]           = df['Invoice_Date'].dt.day
df["Date"]          = pd.to_datetime(df[['Year', 'Month', 'Day']])
df["Hour"]          = df['Invoice_Date'].dt.hour
df["Day_of_Year"]   = df['Invoice_Date'].dt.dayofyear
df["Year_Quarter"]  = df['Year']*10 + df['Quarter']
df["Year_Month"]    = df['Year']*100 + df['Month']
df["Year_Week"]     = df['Year']*100 + df['Week']
df["Year_Day"]      = df['Year']*1000 + df['Invoice_Date'].dt.dayofyear
```
