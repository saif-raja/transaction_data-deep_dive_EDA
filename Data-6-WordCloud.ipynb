```python
import pandas as pd
import numpy as np

pd.options.mode.use_inf_as_na = True
# pd.set_option('')

file = 'data_cleaned\\Data_1-Cleaned.csv'
df = pd.read_csv(file)
df
```

```python
all_text = " ".join(df['Product_Desc'].astype(str).to_list())
# print(all_text)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

ignore_words = ""
ignore_words_list = ignore_words.split()

STOPWORDS.update(ignore_words_list)
wordcloud = WordCloud(background_color="white" , width= 1000 , height= 600 ).generate(all_text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
